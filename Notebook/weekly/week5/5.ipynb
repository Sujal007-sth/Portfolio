```python
import sys

if len(sys.argv) < 2:
    print("Please provide temperature values")
else:
    try:
        temperatures = [float(t) for t in sys.argv[1:]]
        maximum = max(temperatures)
        minimum = min(temperatures)
        average = sum(temperatures) / len(temperatures)
        print(f"Max: {maximum}C, Min: {minimum}C, Mean: {average}C")
    except ValueError:
        print("Please enter valid numbers for temperatures")
```
