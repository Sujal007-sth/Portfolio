```python
# 8. Modify the previous program so that it can process any number of values. The input
# terminates when the user just pressed "Enter" at the prompt rather than entering a value.

temps = []
while True:
    entry = input("Temperature (Enter to stop): ")
    if entry == "":
        break
    value = float(entry.replace("C", ""))
    temps.append(value)

if temps:
    print("Max:", max(temps), "C")
    print("Min:", min(temps), "C")
    average = sum(temps) / len(temps)
    print("Mean:", average, "C")
else:
    print("No data entered")
```
